```python
df['sex'].replace({'female':2, 'male':3}, inplace=True)
df['region'].replace({'southwest':2, 'southeast':3, 'northwest':4, 'northeast':5}, inplace=True)
df['smoker'].replace({'yes':2, 'no':3}, inplace=True)
df
```

### <p style='padding:10px;background-color:#2F4F4F;margin:0;color:#DAA520;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 50px;overflow:hidden;font-weight:500'>Modeling Machine</p>

```python
from sklearn.linear_model import LinearRegression # algorithm of modeling
from sklearn.model_selection import train_test_split # spliting data to two category 
from sklearn import metrics # measure of modeling of machine
```

```python
X = pd.DataFrame(df.drop(['charges'], axis= 1))
y = df['charges'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)

print('X_train:',X_train.shape,'/', 'y_train:',y_train.shape)
print('X_test:',X_test.shape, '/', 'y_test:',y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### <p style='padding:10px;background-color:#2F4F4F;margin:0;color:#DAA520;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 50px;overflow:hidden;font-weight:500'>Metrics</p>

```python
# MAE and MSE and RMSE loser is better

print('Mean Absoult Error: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

#R2_Score greater is better
print('R2_score: ', metrics.r2_score(y_test, y_pred))
```

```python
a = X_test.age
b = y_test
c = X_test.age
d = y_pred

plt.figure(figsize = (5,5))
plt.scatter(a,b, color = 'green')
plt.scatter(c,d, color = 'red')
plt.grid()
plt.xlabel('Age')
plt.ylabel('Y_pred')
plt.title('Age Vs Y_pred')
plt.show() 
```

```python
a = X_test.bmi
b = y_test
c = X_test.bmi
d = y_pred

plt.figure(figsize = (5,5))
plt.scatter(a,b, color = 'green')
plt.scatter(c,d, color = 'red')
plt.grid()
plt.xlabel('BMI')
plt.ylabel('Y_pred')
plt.title('BMI Vs Y_pred')
plt.show()  
```
